```python
dataFrame = dataFrame.drop("transmission",axis=1)
# transmission kısmında sayısal bir değer olmadığı için o kısmı verimizden kaldırdık
```

```python
dataFrame
```

```python
# Model Oluşturmak
y = dataFrame["price"].values # Ulaşmak istediğimiz hedef
x = dataFrame.drop("price",axis=1).values # price harici her şeyi x e ata.
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
# Model eğitimi için
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=10)
# verilerimizi test ve train olarak ayırıyor.
# test_size %30, train_size = %70
#random_state = rastegelelik oranı
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential # Model oluşturmak için
from tensorflow.keras.layers import Dense  # Katmanları eklemek için
```

```python
model = Sequential()
model.add(Dense(12,activation = "relu"))
model.add(Dense(12,activation = "relu"))
model.add(Dense(12,activation = "relu"))
model.add(Dense(12,activation = "relu"))

model.add(Dense(1))

model.compile(optimizer = "adam", loss = "mse")
```

```python
x_train.shape
```

```python
model.fit(x = x_train, y = y_train, validation_data=(x_test,y_test), batch_size = 250 , epochs=8000)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
```

```python
kayipVerisi.head()
```

```python
kayipVerisi.plot()
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)
```
