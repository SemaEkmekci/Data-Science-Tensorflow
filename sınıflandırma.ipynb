```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type", axis = 1).values # x = Type dışındaki bütün veriler
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=25)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
# Bütün x_train verileri 0 ile 1 arasında küçültülüyor.
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units = 30, activation = "relu")) # Kaç kolan varsa nöron (layer) sayısını ona eşitlemek daha mantıklıdır.
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 1, activation = "sigmoid"))

model.compile(loss="binary_crossentropy", optimizer = "adam")
# Sınıflandırma işlemi yaptığımız için loss olarak "binary_crossentropy" kullandık.
```

```python
model.fit(x = x_train, y = y_train, epochs = 700,validation_data=(x_test,y_test), verbose=1)
```

```python
model.history.history
# lossları json formatında görebiliriz. (Sözlük olarak)
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
EarlyStopping = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25)
# Modelin overfitting olmaması için yapılır. Eğer val_loss değeri loss değerinden çok fazla artmaya başladıysa eğitim durur.
```

```python
model.fit(x=x_train, y = y_train, epochs = 700, validation_data = (x_test, y_test), verbose = 1, callbacks=[EarlyStopping] )
```

```python
modelKaybi = pd.DataFrame(model.history.history)
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units = 30, activation = "relu")) # Kaç kolan varsa nöron (layer) sayısını ona eşitlemek daha mantıklıdır.
model.add(Dropout(0.6)) # 1'den küçük değerler verilmelidir.
model.add(Dense(units = 15, activation = "relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 1, activation = "sigmoid"))

model.compile(loss="binary_crossentropy", optimizer = "adam")
# Sınıflandırma işlemi yaptığımız için loss olarak "binary_crossentropy" kullandık.
```

```python
model.fit(x=x_train, y = y_train, epochs = 700, validation_data = (x_test, y_test), verbose = 1, callbacks=[EarlyStopping] )
```

```python
kayipDf = pd.DataFrame(model.history.history)
```

```python
kayipDf.plot()
```

```python
tahminlerimiz = model.predict(x_test)
predicted_classes = tahminlerimiz.argmax(axis=-1)
```
