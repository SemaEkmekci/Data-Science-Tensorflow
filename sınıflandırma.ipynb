```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type", axis = 1).values # x = Type dÄ±ÅŸÄ±ndaki bÃ¼tÃ¼n veriler
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=25)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
# BÃ¼tÃ¼n x_train verileri 0 ile 1 arasÄ±nda kÃ¼Ã§Ã¼ltÃ¼lÃ¼yor.
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units = 30, activation = "relu")) # KaÃ§ kolan varsa nÃ¶ron (layer) sayÄ±sÄ±nÄ± ona eÅŸitlemek daha mantÄ±klÄ±dÄ±r.
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 1, activation = "sigmoid"))

model.compile(loss="binary_crossentropy", optimizer = "adam")
# SÄ±nÄ±flandÄ±rma iÅŸlemi yaptÄ±ÄŸÄ±mÄ±z iÃ§in loss olarak "binary_crossentropy" kullandÄ±k.
```

```python
model.fit(x = x_train, y = y_train, epochs = 700,validation_data=(x_test,y_test), verbose=1)
```

```python
model.history.history
# losslarÄ± json formatÄ±nda gÃ¶rebiliriz. (SÃ¶zlÃ¼k olarak)
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units = 30, activation = "relu")) # KaÃ§ kolan varsa nÃ¶ron (layer) sayÄ±sÄ±nÄ± ona eÅŸitlemek daha mantÄ±klÄ±dÄ±r.
model.add(Dropout(0.6)) # 1'den kÃ¼Ã§Ã¼k deÄŸerler verilmelidir.
model.add(Dense(units = 15, activation = "relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 1, activation = "sigmoid"))

model.compile(loss="binary_crossentropy", optimizer = "adam")
# SÄ±nÄ±flandÄ±rma iÅŸlemi yaptÄ±ÄŸÄ±mÄ±z iÃ§in loss olarak "binary_crossentropy" kullandÄ±k.
```

```python
model.fit(x=x_train, y = y_train, epochs = 700, validation_data = (x_test, y_test), verbose = 1, callbacks=[EarlyStopping] )
```

```python
kayipDf = pd.DataFrame(model.history.history)
```
